```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.impute import SimpleImputer

class AgeImputer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        imputer = SimpleImputer(strategy='mean')
        X['Age'] = imputer.fit_transform(X[['Age']])
        return X
```

```python
from sklearn.preprocessing import OneHotEncoder

class FeatureEncoder(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        encoder = OneHotEncoder()

        X['Embarked'] = X['Embarked'].fillna('Missing')
        embarked_matrix = encoder.fit_transform(X[['Embarked']]).toarray()
        embarked_columns = [str(category) for category in encoder.categories_[0]]
        for i, col in enumerate(embarked_columns):
            X[col] = embarked_matrix[:, i]

        # Fill missing values for 'Sex'
        X['Sex'] = X['Sex'].fillna('Missing')
        sex_matrix = encoder.fit_transform(X[['Sex']]).toarray()
        sex_columns = [str(category) for category in encoder.categories_[0]]
        for i, col in enumerate(sex_columns):
            X[col] = sex_matrix[:, i]

        return X

```

```python
# Correcting the FeatureDropper class
class FeatureDropper(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        return X.drop(['Embarked', 'Name', 'Ticket', 'Sex', 'Cabin', 'Missing'], axis=1, errors='ignore')
```

```python
from sklearn.pipeline import Pipeline

# Defining the pipeline with corrected transformations
pipeline = Pipeline([
    ('age_imputer', AgeImputer()),
    ('feature_encoder', FeatureEncoder()),
    ('feature_dropper', FeatureDropper())
])
```
